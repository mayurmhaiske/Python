4.	Inheritance (The super() function):

o	Task: Create a subclass named GoldenRetriever that inherits from Dog. Override the bark(self) method to print "Happy Woof!" instead of "Woof!".
    
o	Action: Create an instance of GoldenRetriever and call its bark() method.


```python
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        print(f"Woof! My name is {self.name}")

class GoldenRetriever(dog):
    def bark(self):
        print("Happy Woof!")

gr_dog = GoldenRetriever("Sunny")

gr_dog.bark()

```

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.__age = age  # Private attribute

    def get_age(self):
        return self.__age  # Public method to access private attribute

# Create an instance of Dog
my_dog = Dog("Buddy", 5)

# Try to access __age directly (will cause an AttributeError)
try:
    print(my_dog.__age)
except AttributeError:
    print("Error: Cannot access private attribute __age directly.")

# Access __age correctly using the public method
print("Dog's age is:", my_dog.get_age())
```
