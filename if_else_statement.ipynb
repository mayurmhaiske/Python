```python
value_if_true if condition else value_if_false
```

```python
age = 19
status = "Adult" if age>= 18 else "minor"
print(status)
```

```python
score = int(input())

if score >= 90:
    print("A")
else:
    if score >= 70:
        print("B")
    else:
        print("C")
```

```python
score = 76
grade = "A" if score >=90 else "B" if score >= 75 else "C"
print(grade)
```

```python
from datetime import datetime

current_hour = datetime.now().hour

greetings = (
     "good morning" if current_hour < 12 else
     "good afternoon" if current_hour < 18 else
     "good evening" if current_hour < 21 else
     "good night"
)

print(greetings)
```

### Match case


```python
day = "sunday"

match day:
    case "monday":
       print("this is start of week")
    case "tuesday":
         print("this is a second day of the week")
    case "wednesday":
         print("this is mid of the week")
    case _ :
        print("its just a another day")
        
```
